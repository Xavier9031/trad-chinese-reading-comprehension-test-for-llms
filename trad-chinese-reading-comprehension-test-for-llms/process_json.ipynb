```python
import pandas as pd

# Load the dataset
file_path = 'AI.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Preparing the data in the specified JSON format
json_data = []

for index, row in data.iterrows():
    # Combine the article and question
    input_text = str(row['文章']) + "\n問題：" + str(row['問題']) + "\n選項1：" + str(row['選項1']) + "\n選項2：" + str(row['選項2']) + "\n選項3：" + str(row['選項3']) + "\n選項4：" + str(row['選項4'])

    # Format into the desired structure
    json_entry = {
        "instruction": "請根據以下輸入內容詳細思考後，回答選擇題，並只以阿拉伯數字回答：\n",
        "input": input_text,
        "output": str(row['正確答案'])
    }
    json_data.append(json_entry)

# Convert the list to JSON
import json
json_output = json.dumps(json_data, ensure_ascii=False, indent=4)

# Save the JSON file
with open('AI.json', 'w') as outfile:
    outfile.write(json_output)


```

```python
import pandas as pd

# Load the dataset
file_path = 'AI1000.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Preparing the data in the specified JSON format
json_data = []

for index, row in data.iterrows():
    # Combine the article and question
    input_text = str(row['文章']) + "\n問題：" + str(row['問題']) + "\n選項1：" + str(row['選項1']) + "\n選項2：" + str(row['選項2']) + "\n選項3：" + str(row['選項3']) + "\n選項4：" + str(row['選項4'])

    # Format into the desired structure
    json_entry = {
        "instruction": "請根據以下輸入內容詳細思考後，回答選擇題，並只以阿拉伯數字回答：\n",
        "input": input_text,
        "id": str(row['題號'])
    }
    json_data.append(json_entry)

# Convert the list to JSON
import json
json_output = json.dumps(json_data, ensure_ascii=False, indent=4)

# Save the JSON file
with open('AI1000.json', 'w') as outfile:
    outfile.write(json_output)
```
